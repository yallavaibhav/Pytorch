## Tensor datatypes
Note: Tensor datatypes is one of the 3 big errors u will run into with Pytorcha nd DL
1. Tensors not right data types
2. Tensors not right shape
3. Tensors not on the right device


```python
Float_32_tensor = torch.tensor([3.0,6.0,9.0], 
                               dtype=None, # what datatype is the tensor (eg. flaot32 or float 16)
                               device = None, # What device is your tensor on
                               requires_grad=False) # wheather or not to track gradients with this tensors operations
Float_32_tensor
```

```python
Float_32_tensor.dtype
```

```python
Float_16_tensor = Float_32_tensor.type(torch.float16)
Float_16_tensor
```

```python
Float_16_tensor.dtype
```

```python
Float_32_tensor * Float_16_tensor
```

## Getting information from tensors
1. Tensors not right data types - to do get datatype, can use tensor.dtype
2. Tensors not right shape - to get shape from a tensor, can use tensor.shape
3. Tensors not on the right device-to get device from a tensor, can use tensor.device

```python
# Create a tensor
some_tensor = torch.rand(3,4)
some_tensor
```

```python
print(some_tensor)
print("Shape of the some_tensor is",some_tensor.shape)
print("Data type of the some_tensor is ",some_tensor.dtype)
print("Device of the some_tensor is ",some_tensor.device)
```
